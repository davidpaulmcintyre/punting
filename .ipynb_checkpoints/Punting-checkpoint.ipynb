```python
# def get_punter_name(desc):
#     desc field contains text such as "(13:59) J.Ryan punts 57 yards..."
#     print(desc)
    
#     if 'punts' in desc:
#         desc_words = desc.split()
#         ix_punts = desc_words.index('punts')
#         ix_punter_name = ix_punts - 1
#         return desc_words[ix_punter_name] if ix_punter_name > 0 else ''
#     return 'unknown'

lefties = ['R.Allen', 'C.Bojorquez', 'D.Colquitt', 'M.Haack', 'C.Hanson', 'C.Jones', 'D.Jones', 
           'J.Kapinos', 'J.Locke', 'Z.Mesko', 'M.Palardy', 'G.Pakulak', 'M.Palardy', 'D.Sepulveda', 'T.Way', 
           'B.Wing', ]

'''
righties = ['Nick.Harris' 'M.Koenen' 'T.Morstead' 'S.Koch' 'M.Scifres'
 'C.Kluwe' 'D.Zastudil'   'S.Rocca' 'A.Lee' 'P.McAfee'
 'S.Lechler' 'J.Baker' 'K.Huber' 'M.McBriar' 'B.Fields' 'J.Ryan'
 'B.Maynard' 'B.Graham'  'C.Bryan' 'T.Masthay' 'M.Dodge'
 'B.Moorman' 'H.Smith'  'R.Hodges' 'B.Kern' 'N.Harris'
 'S.Weatherford' 'A.Podlesh' 'T.Conley'  'J.Hekker' 'R.Malone'
 'unknown' 'S.Powell' 'S.Lanning'   'B.Anger'
 'B.Colquitt'  'M.King' 'B.Nortman' "P.O'Donnell"  
 'D.Butler'   'J.Schum' 'R.Quigley'  'S.Martin' 'R.Dixon'
 'B.Pinion' 'L.Edwards' 'J.Berry' 'D.Kaser', 'C.Schmidt', 'C.Hentrich', M.McBriar, 'D.Johnson', 'M.Turk', 'J.Feagles',
 'M.Berger', 'S.Paulescu', C.Barth, M.Dodge, J.Bidwell, R.Malone, C.Henry, R.Donahue, M.Bosher, B.Hartmann, R.McGee,
 R.McGee, S.Powell, M.Darr, J.Schum, R.Dixon, M.Wile, K.Redfern, R.Sanchez, J.Vogel,  'C.Johnston', 'L.Cooke', 'T.Daniel',
 'M.Dickson', 'J.Scott', 'J.Townsend', 'C.Wadman'
 
 '''
# S.Suisham did some non-pooch punting as well
not_punters = ['B.Roethlisberger', 'T.Brady', 'S.Janikowski', 'P.Dawson', 'B.Cundiff', 'D.Carpenter' 'C.Sturgis', 'D.Bailey', 'C.Catanzaro', 'G.Gano', 'R.Succop', 'M.Prater' ]
# aussie punters: C.Bryan, Sav Rocca, 

# punt_fair_catch
# fumble_forced
# punt_downed
# fumble_lost
# penalty_type
def muffed_punt(desc):
#     desc field conventionally contains term 'for no gain.' if fumbled on catch
    return 1 if 'MUFFS' in desc else 0 

def is_lefty(p):
    return 1 if p in lefties else 0
  
def faircatch(desc):
#     desc field conventionally contains term 'fair catch'  
    return 'fair catch' in desc

def not_fielded(desc):
#     how to tell if punt desc returner didnt attempt to field ball
    pass

def blocked(desc):
    pass
# replace with Punt_Blocked
#     return 'punt is BLOCKED' in desc

def roughing_penalty(desc):
    return 'Roughing the Kicker' in desc

# excl plays where team punts from field goal formation or qb punts

# df = pd.read_csv('nflplaybyplay2009to2016/NFL Play by Play 2009-2016 (v3).csv')
# print('start')
# df = pd.read_csv('nflplaybyplay2009to2016/NFL Play by Play 2009-2017 (v4).csv') 
    
# print(output.head(50))
# print(df_punts.punter_player_name.unique())
# df_punts['Punter'] = df_punts.desc.apply(get_punter_name)
# df_punts['Muffed'] = df_punts.desc.apply(muffed_punt)
# print(df_punts.PuntResult.unique()) ['Clean' 'Blocked']
# list_punters = sorted(df_punts.Punter.unique())
# print(list_punters)
# list_vol4 = []
# columns = ['Date', 'RecFumbTeam', 'Punter', 'Muffed', 'HomeTeam', 'AwayTeam', 'desc']
# list_vol3 = ['A.Lee', 'A.Podlesh', 'B.Anger', 'B.Colquitt', 'B.Fields', 'B.Graham', 'B.Kern', 'B.Maynard', 'B.Moorman', 'B.Nortman', 'B.Pinion', 'B.Wing', 'C.Bryan', 'C.Hanson', 'C.Jones', 'C.Kluwe', 'D.Butler', 'D.Colquitt', 'D.Jones', 'D.Kaser', 'D.Sepulveda', 'D.Zastudil', 'G.Pakulak', 'H.Smith', 'J.Baker', 'J.Berry', 'J.Hekker', 'J.Kapinos', 'J.Locke', 'J.Ryan', 'J.Schum', 'K.Huber', 'L.Edwards', 'M.Dodge', 'M.King', 'M.Koenen', 'M.McBriar', 'M.Scifres', 'N.Harris', 'Nick.Harris', 'P.McAfee', "P.O'Donnell", 'R.Allen', 'R.Dixon', 'R.Hodges', 'R.Malone', 'R.Quigley', 'S.Koch', 'S.Lanning', 'S.Lechler', 'S.Martin', 'S.Powell', 'S.Rocca', 'S.Weatherford', 'T.Conley', 'T.Masthay', 'T.Morstead', 'T.Way', 'Z.Mesko', 'unknown']
# for p in list_punters:
#     if p not in list_vol3:
#         list_vol4.append(p)
# df_punts[columns].to_csv('fumbles.csv', index=False)
# "for no gain" indicates muffed punt 


def get_data(): 
#     use latest volume of data
    df = pd.read_csv('nflplaybyplay2009to2016/NFL Play by Play 2009-2018 (v5).csv') 
#     include only punts. exclude pooch punts by QBs and placekickers
#     excluding non-punters and nan removes 329 rows from righty count
    df_punts =  df[(df.punt_attempt == 1) & (~df.punter_player_name.isin(not_punters)) & (df.punter_player_name.notnull())]
    return df_punts 


def compare_lefty_right(df_punts):
#     statistical comparison of L vs R foot fumbling rates
    punts_count = df_punts.shape[0]
    print('total punts: ', punts_count) 
    df_punts['lefty'] = df_punts.punter_player_name.apply(is_lefty) 
    df_punts['muffed'] = df_punts.desc.apply(muffed_punt)  
    print('corr: ', df_punts.muffed.corr(df_punts.lefty)) 
    crosstab = pd.crosstab(index=df_punts.muffed, columns=df_punts.lefty)
    print('crosstab: ')
    print(crosstab)
    punt_count_righty = df_punts[df_punts.lefty == 0].shape[0]
    punt_count_lefty = df_punts[df_punts.lefty == 1].shape[0]
    by_lefties = punt_count_lefty / punts_count * 100
#     print('Pct of punts by lefties: {0}' % by_lefties )
    print("Pct of punts by lefties: {:.2f}".format(by_lefties))
#     print('{0} and {1}'.format('Geeks', 'Portal')) 
    muffs_righty = df_punts[(df_punts.lefty == 0) & (df_punts.muffed == 1)].lefty.count()
    muffs_lefty = df_punts[(df_punts.lefty == 1) & (df_punts.muffed == 1)].lefty.count()
    muff_rate_righty = muffs_righty / punt_count_righty * 100
    muff_rate_lefty = muffs_lefty / punt_count_lefty * 100 
    print("Muff rate: %(r)s%% %(l)s%%" % {'r': round(muff_rate_righty, 2), 'l': round(muff_rate_lefty, 2)})
    stat, p, dof, expected = stats.chi2_contingency(crosstab)
    print('p ',  p)
    
def compare_teams(df):
    pass

def compare_punters(df): 
    df['lefty'] = df_punts.punter_player_name.apply(is_lefty) 
    df['muffed'] = df_punts.desc.apply(muffed_punt) 
#     columns = ['punter_player_name', 'muffed', 'lefty']
    columns_groupby = ['punter_player_name', 'muffed', 'lefty']
    df = df[columns_groupby]  
    df_grouped = df.groupby(columns_groupby).agg({'muffed': 'count'})   
#     create separate df for aggregation of counts of punts per punter
    df_subtotals_count = df_grouped.groupby(level='punter_player_name').apply(lambda x: x.sum()) 
    df_subtotals_count = df_subtotals_count.rename(index=str, columns={'muffed': 'total_punts'})
    df_subtotals_count = df_subtotals_count.reset_index() 
#     create separate df for aggregation of counts of punter muffed pct
    df_subtotals_percent = df_grouped.groupby(level='punter_player_name').apply(lambda x: 100 * x / x.sum()) 
    
    df_subtotals_percent = df_subtotals_percent.reset_index(level='punter_player_name')
    df_subtotals_percent = df_subtotals_percent.rename(index=str, columns={'muffed': 'muffed_percent'})
    df_subtotals_percent = df_subtotals_percent.reset_index()
    df_subtotals_percent.muffed = df_subtotals_percent.muffed.astype(int)
    df_subtotals_percent = df_subtotals_percent[df_subtotals_percent.muffed == 1]
    df_subtotals_percent = df_subtotals_percent.drop('muffed', axis=1) 
    df_merged = pd.merge(df_subtotals_percent, df_subtotals_count, on='punter_player_name')
    df_merged = df_merged[['punter_player_name', 'lefty', 'total_punts', 'muffed_percent']]
    df_merged.muffed_percent = df_merged.muffed_percent.map('{:,.2f}'.format)
#     exclude punters below minimum threshold
    df_merged = df_merged[df_merged.total_punts > 300]
    output_sorted = df_merged.sort_values(['muffed_percent'], ascending=False)
    output_sorted = output_sorted.reset_index(drop=True) 
    return output_sorted.head(20)
```
